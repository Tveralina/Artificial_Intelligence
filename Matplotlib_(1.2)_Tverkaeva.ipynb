```
from PIL import Image
from pylab import *

# Чтение изображения в массив
im = array(Image.open('/content/drive/MyDrive/Colab Notebooks/DSC5900.jpg'))

# Поместить на график изображение
imshow(im)

# Несколько точек
x = [100,100,400,400]
y = [200,500,200,500]

# Нанести точки в виде красных звездочек
plot(x,y,'r*')

# Нарисовать желтый отрезок, соединяющих первые две точки
plot(x[:2],y[:2],'y')

# Нарисовать диагональный белый отрезок
plot(x[1:3],y[1:3],'w')

# Добавление заголовка и показ графика
title('Plotting:"View from Moscow City"')
show()
```

```
# Чтение изображения в массив
im_wb = array(Image.open('/content/drive/MyDrive/Colab Notebooks/DSC5900.jpg').convert('L'))

# Создание нового рисунка
figure()

# Не использовать цвета
gray()

# Показать изолинии относительно левого верхнего угла
contour(im_wb, origin='image')
axis('equal')
axis('off')
```

```
# Построение гистограммы полутонового изображения
hist(im_wb.flatten(),256)
show()
```
